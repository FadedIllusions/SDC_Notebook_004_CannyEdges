```python
# Import Needed Packages
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
```

```python
# Read In Image And Display
image = mpimg.imread('exit_ramp.jpg')
plt.imshow(image)
```

```python
# Convert Image To Grayscale And Display
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
# Gaussian Blur Image With A Kernel Size Of 3
kernel_size = 5
blur_gray = cv2.GaussianBlur(gray,(kernel_size, kernel_size), 0)
```

```python
# Define Gaussian Parameters And Run Gaussian Blur Function
low_threshold = 1
high_threshold = 200
edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
```

```python
# Display Gaussian Image
plt.imshow(edges, cmap="Greys_r")
```
